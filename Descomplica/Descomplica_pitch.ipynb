```python
import pandas as pd

#df = pd.read_csv("/content/World CO2 Emission Data.csv")
df_total = pd.read_csv("/content/drive/MyDrive/Descomplica_pitch/World CO2 Emission Data.csv")

#df_total.head(15)
```

```python
porcentagem_treinamento = 0.8

df_train = []
df_val = []

for i in range(0, len(df_train_val)):
  train_size = int(len(df_train_val[i]) * porcentagem_treinamento)
  df_train.append(df_train_val[i][:train_size])
  df_val.append(df_train_val[i][train_size:])

print(df_train[0].tail(10))
print(df_val[0].head(10))
print(len(df_train[12]),len(df_val[12]))
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Definir o número de meses anteriores a serem considerados
#k = 24
k_max=16
k_Germany = len(df_train[12])
k_usage = []

rmse_train_vector = []
rmse_val_vector = []

for i in range(0,len(df_train)):
  rmse_train_vector.append([])
  rmse_val_vector.append([])

best_model = []
best_k = []
best_rmse_val = []
best_mape_val = []

controle = 0
for j in range(0,len(df_train)):
  if (len(df_train[j]) <= k_Germany) & (k_max > k_Germany):
    k_usage.append(k_Germany)
  else:
    k_usage.append(k_max)
  #print(k_usage[j])
  print(df_train[j])
  for k in range(1, k_usage[j]):

    # Criar matrizes X e y para treinamento
    X_train = []
    y_train = []
    for i in range(k, len(df_train[j])):
      X_train.append(df_train[j]['value'][i-k:i])
      y_train.append(df_train[j]['value'][i])
    X_train = np.array(X_train)
    y_train = np.array(y_train)

    print("Conjunto X: ", X_train)
    print("Conjunto y: ", y_train)

    # Treinar o modelo de regressão linear
    model = LinearRegression()
    model.fit(X_train, y_train)

    # Calcular o RMSE do conjunto de treinamento
    y_pred_train = model.predict(X_train)
    rmse_train = np.sqrt(np.mean((y_train - y_pred_train)**2))

    # prompt: Preciso predizer o o conjunto de validação de df_val agora

    Ajuste_dados_previos_val = True
    # Ajusta o vetor de validação para o modelo
    if (Ajuste_dados_previos_val):
      df_val_modelo = pd.concat([df_train[j].iloc[-k:], df_val[j]])
      df_val[j] = pd.concat([df_train[j].iloc[-k:], df_val[j]])
    else:
      df_val_modelo = df_val[j]

    # Criar matrizes X e y para validação
    X_val = []
    y_val = []
    for i in range(k, len(df_val_modelo)):
      X_val.append(df_val_modelo['value'][i-k:i])
    X_val = np.array(X_val)

    y_val.append(df_val_modelo['value'][k:])
    y_val = np.array(y_val)

    # Predizer o conjunto de validação
    y_pred_val = model.predict(X_val)

    # Calcular o RMSE do conjunto de validação
    rmse_val = np.sqrt(np.mean((y_val - y_pred_val)**2))

    # Calculate the mean absolute percentage error (MAPE) for the test set
    mape_val = np.mean(np.abs((y_val - y_pred_val) / y_val)) * 100

    if k==1:
      best_model.append(model)
      best_rmse_val.append(rmse_val)
      best_k.append(k)
      best_mape_val.append(mape_val)
    elif best_rmse_val[j] > rmse_val:
      best_model[j] = model
      best_rmse_val[j] = rmse_val
      best_k[j] = k
      best_mape_val[j] = mape_val
    elif k==8:
      best_model[j] = model
      best_rmse_val[j] = rmse_val
      best_k[j] = k
      best_mape_val[j] = mape_val

    if j==0:
      controle = controle + 1

    #print('RMSE do conjunto de validação:', rmse_val)
    rmse_train_vector[j].append(rmse_train)
    rmse_val_vector[j].append(rmse_val)
  print(len(X_val))
```
