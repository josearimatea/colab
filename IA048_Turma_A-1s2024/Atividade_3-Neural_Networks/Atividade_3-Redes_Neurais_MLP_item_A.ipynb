```python
# prompt: Leia o arquivo npz no caminho: /content/drive/MyDrive/Classroom/IA048 Turma A - 1s2024/Atividade_3/bloodmnist_64.npz

import numpy as np
data = np.load('/content/drive/MyDrive/Classroom/IA048_Turma_A-1s2024/Atividade_3/bloodmnist_64.npz')
print("Chaves disponíveis no arquivo:", data.files)

# Load the arrays from the NPZ file
X_train = data['train_images']
y_train = data['train_labels']
X_val = data['val_images']
y_val = data['val_labels']
X_test = data['test_images']
y_test = data['test_labels']
```

```python
# prompt: y_train está como array de vários arrays de um único valor, mas quero que fique como um único array de valores

y_train = np.concatenate(y_train)
# X_train = np.concatenate(X_train)
```

```python
import numpy as np
from sklearn.metrics import confusion_matrix, accuracy_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ModelCheckpoint

# Função para criar o modelo Keras
def create_model():
    model = Sequential()
    model.add(Flatten(input_shape=(64, 64, 3)))
    model.add(Dense(128, activation='relu'))
    model.add(Dense(32, activation='softmax'))
    model.compile(optimizer=Adam(), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model

# Crie o modelo Keras
model = create_model()

# Defina o callback ModelCheckpoint para salvar o melhor modelo
checkpoint_callback = ModelCheckpoint(filepath='best_model.h5', monitor='val_accuracy', save_best_only=True, verbose=1)

# Treine o modelo nos dados de treinamento e validação, usando o callback ModelCheckpoint
model.fit(X_train, y_train, epochs=10, validation_data=(X_val, y_val), callbacks=[checkpoint_callback])
```

```python
from tensorflow.keras.models import load_model

# Carregue o melhor modelo salvo
best_model = load_model('best_model.h5')

# Avalie o melhor modelo nos dados de teste
y_test_pred = best_model.predict(X_test)
y_test_pred = np.argmax(y_test_pred, axis=1)
accuracy_test = accuracy_score(y_test, y_test_pred)
confusion_matrix_test = confusion_matrix(y_test, y_test_pred)

# Imprima os resultados
print('Acurácia nos dados de teste:', accuracy_test)
print('Matriz de confusão nos dados de teste:')
print(confusion_matrix_test)
```
