```python
import numpy as np
data = np.load('/content/drive/MyDrive/Colab_notebooks/IA048_Turma_A-1s2024/Atividade_3/bloodmnist_64.npz')
print("Chaves disponíveis no arquivo:", data.files)

# Load the arrays from the NPZ file
X_train = data['train_images']
y_train = data['train_labels']
X_val = data['val_images']
y_val = data['val_labels']
X_test = data['test_images']
y_test = data['test_labels']
```

```python

import tensorflow as tf
from tensorflow.keras.callbacks import ModelCheckpointfrom tensorflow.keras.models import load_model

best_model = []
accuracy = []

for n_kernel in range(1,32)
  # Define the CNN model
  model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(n_kernel, (3, 3), activation='relu', input_shape=(64, 64, 3)),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(32, activation='softmax')
  ])

  # Compile the model
  model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

  # Defina o callback ModelCheckpoint para salvar o melhor modelo
  checkpoint_callback = ModelCheckpoint(filepath='best_model.h5', monitor='val_accuracy', save_best_only=True, verbose=1)

  # Train the model
  model.fit(X_train, y_train, epochs=10, validation_data=(X_val, y_val), callbacks=[checkpoint_callback])

  # Carregue o melhor modelo salvo
  best_model.append(load_model('best_model.h5'))
```
